```python
names=['chlorhexidine', 'cyntomycin', 'afobazol']
counts=[15, 18, 7]
```

```python
def create_medications(names, counts):
    return pd.Series(index=names,data=counts)

medications=create_medications(names, counts)

display(medications)
```

```python
def get_percent(medications, name):
    all_sum=medications[:].sum()
    medications[name]
    return(medications[name]/all_sum*100)
    
print(get_percent(medications, "chlorhexidine"))
```

```python
income = [478, 512, 196]
expenses = [156, 130, 270]
years = [2018, 2019, 2020]

def create_companyDF(income, expenses, years):
    res_df=pd.DataFrame({'income':income,'expenses':expenses},index=years)
    return res_df

df=create_companyDF(incomes, expenses, years)    
display(df)

#print(df.index)

def get_profit(df, year):
    if year in df.index :
        return df.loc[year,'income']-df.loc[year,'expenses']
    else:
        return None

print(get_profit(df,2018))
```
